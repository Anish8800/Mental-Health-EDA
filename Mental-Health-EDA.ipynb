# **Project Summary -**

This project focuses on understanding patterns, risk factors, and workplace influences associated with mental health among employees in the technology industry. The dataset used in this project consists of 1259 responses collected through an online survey regarding attitudes, prevalence, and workplace support related to mental health conditions.

The goal of the project is not to build predictive models but to explore, visualize, and derive insights from the data. Exploratory Data Analysis helps us understand relationships between variables such as age, gender, geography, family history, employer support policies, and the likelihood of an employee seeking treatment for mental health conditions.

The dataset contains both numerical and categorical variables. Important fields include age, gender, country, employment type, family mental health history, willingness to seek treatment, ease of taking leave, perceived consequences at work, and availability of employer-provided benefits. Before performing analysis, significant data cleaning was required. Missing values were handled, duplicate rows were checked, inconsistent age values were removed, and categorical responses were standardized.

Interesting insights emerged during the analysis. Approximately 50% of respondents reported receiving mental health treatment, indicating that mental health issues are fairly prevalent in the technology workforce. Around 39% reported a family history of mental health problems, reinforcing the role of hereditary factors. It was also observed that employees who believed mental health issues could negatively affect their careers were less likely to seek help, highlighting the impact of stigma.

Workplace culture plays an important role. Respondents who reported supportive employer policies such as mental health benefits, flexible leave for mental health, and open discussions with supervisors were more likely to seek treatment when needed. On the contrary, lack of benefits and fear of consequences discouraged disclosure and treatment seeking. Another key result was that employees who frequently experienced mental health interference in work were more willing to access support.

Age patterns showed concentration between 20 and 40 years, indicating that working-age adults are a major affected group. There were extreme invalid age values which were removed during cleaning. Gender diversity was present, but most responses came from males. The majority of responses originated from the United States, followed by other Western countries.

The findings of this EDA highlight that mental health remains a serious concern in the technology industry. Stigma, inadequate workplace support, and fear of consequences contribute to untreated cases. However, organizations with open culture and supportive policies show positive impact on employees’ willingness to seek help.

This project demonstrates how data exploration can support policy decisions. Companies may use such analysis to design awareness programs, expand benefits, encourage disclosure without fear, and create psychologically safe workplaces. The project reinforces the idea that addressing mental health is not only a personal necessity but also improves productivity, morale, and long-term employee well-being.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# 1. Remove invalid ages
df = df[(df['Age'] >= 18) & (df['Age'] <= 80)]

# 2. Handle missing values
df['self_employed'] = df['self_employed'].fillna('No')
df['work_interfere'] = df['work_interfere'].fillna('Unknown')

# 3. Remove unnecessary column if exists
if 'comments' in df.columns:
    df.drop(columns=['comments'], inplace=True)

# 4. Drop duplicates
df.drop_duplicates(inplace=True)

df.head()
```

### What all manipulations have you done and insights you found?

Following manipulations were performed:

Removed invalid age values below 18 and above 80

Filled missing values in:

self_employed → No

work_interfere → Unknown

Dropped comments column as it is free-text and not useful for analysis

Removed duplicate records

Converted categorical text responses to consistent formatting (upper/lower-case handled automatically by pandas)

Insights Found:

Most respondents belong to the age group 20–40 years

Majority work in technology companies

Family history of mental illness is common

Nearly half have sought treatment

Many employees report that mental issues interfere with work
